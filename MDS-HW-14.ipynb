```python
import numpy as np 

x = np.array([0,1,2,3,4,5]) 
y = np.array([-100,-90,-76,-52,-12,50]) 

n = np.size(x) 

_b_divisor = (n*(np.sum(x*x) - np.sum(x)**2))
alfa_b = (np.sum(x*x) * np.sum(x) - n*np.sum(x*x*x)) / _b_divisor
betta_b = (n*np.sum(x*y) - np.sum(y)*np.sum(x)) / _b_divisor

alfa_c = (np.sum(x*x) + alfa_b * np.sum(x)) / -n
betta_c = (np.sum(y) - betta_b * np.sum(x)) / n

a = (np.sum(y*x*x) - betta_b * np.sum(x*x*x) - betta_c * np.sum(x*x)) / \
    (np.sum(x*x*x*x) + alfa_b*np.sum(x*x*x) + alfa_c*np.sum(x*x))

b = a*alfa_b + betta_b

c = a*alfa_c + betta_c
```

```python
print(alfa_b)
print(betta_b)
print()
print(alfa_c)
print(betta_c)
print()
print(a)
print(b)
print(c)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.scatter(x, y, color = "m", marker = "o", s = 30) 

y_predicted = a*x*x + b*x + c
plt.plot(x, y_predicted, color = "g", marker = "x")

plt.gcf().set_size_inches(15, 15)
plt.show()
```
